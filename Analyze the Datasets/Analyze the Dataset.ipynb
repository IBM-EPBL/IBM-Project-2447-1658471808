```python
# Importing Necessary Libraries
import numpy as np # linear algebra
import pandas as pd # data processing, csv file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
```

```python
path = "data/Wind turbine.csv"

df = pd.read_csv('E:\study\Project Design & Planning\Data collection\Download dataset\Wind turbine.csv')

df.rename(columns={'Date/Time':'Time','LV ActivePower (KW)':'ActivePower(KW)',"Wind Speed (m/s)":"WindSpeed(m/s)","Wind Direction ()":"WindDirecion"},inplace=True)
```

```python
#Data Description and Visualizing

#Plotting the pair plot, each variable in the data set is plotted with all other variables

sns.pairplot(df)

#Plotting Correlation between the variables

plt.figure(figsize=(10, 8)) 
corr= df.corr()
ax = sns.heatmap(corr, vmin = -1, vmax = 1, annot = True) 
bottom, top = ax.get_ylim() 
ax.set_ylim(bottom + 0.5, top - 0.5) 
plt.show()

#Print the numerical form for the correlation 
print(corr)
```
